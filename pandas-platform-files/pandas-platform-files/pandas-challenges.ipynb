Read the `vehicles.csv` dataframe and store it in a variable called `cars`:

Drop the column "Combined MPG"

```python
# code here
cars.drop(columns=['Combined MPG'])
```

Change column names so that there are no names with spaces or weird special characters:

```python
# code here
cars = cars.rename(columns = {
            'Engine Displacement' : 'EngineDisplacement',
            'Vehicle Class' : 'VehicleClass',
            'Fuel Type' : 'FuelType',
            'Fuel Barrels/Year' : 'FuelBarrelsPerYear',
            'City MPG' : 'CityMPG',
            'Highway MPG' : 'HighwayMPG',
            'CO2 Emission Grams/Mile' : 'CO2EmissionGramsPerMile',
            'Fuel Cost/Year' : 'FuelCostPerYear'
})
cars
```

What brand has the most cars?

```python
# code here

no_of_cars = cars.Make.value_counts().reset_index()
no_of_cars[no_of_cars.Make == no_of_cars.Make.max()]
```

What brand has the worse CO2 Emissions on average?

```python
# code here
co2 = cars.groupby('Make')['CO2EmissionGramsPerMile'].mean().reset_index()
co2.columns = ['brand', 'avgco2']
co2[co2.avgco2 == co2.avgco2.max()]
```

Which brands are more environment friendly?

```python
# code here
co2[co2.avgco2 == co2.avgco2.min()]
```

Create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
# code here
# cars.Year.min()
#source to https://www.youtube.com/watch?v=cgpiFkpBLyY

# cut command creates equispaced bins but frequency of samples is unequal in each bin
# qcut command creates unequal size bins but frequency of samples is equal in each bin.

decades = [1980, 1990, 2000, 2010]
cut_bins = [1983, 1989, 1999, 2009, 2017]
cars['Decades'] = pd.cut(cars['Year'], bins = cut_bins, labels=decades)
cars.head()
```

Did cars consume more gas in the eighties?

```python
# code here
cars.groupby(['Decades'])['FuelBarrelsPerYear'].mean()
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
# code here
# in the next notebook, we will learn how we can clean the "Trans" column
cars.groupby(['Transmission'])['FuelBarrelsPerYear'].mean()
```
